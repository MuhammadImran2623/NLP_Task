```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.text import Tokenizer
```

* We need to convert the text data  into a list of strings before Tokenizing
* `Tokenizer` is designed to process multiple texts (such as sentences or documents) simultaneously, and it expects the input to be an iterable (like a list) where each element is a separate text string.
## Key Functions of Tokenizer:
* fit_on_texts(texts): This method updates the internal vocabulary of the tokenizer based on the list of texts.
* texts_to_sequences(texts): Converts texts into sequences of integers.


```python
texts = df1['Text'].tolist()
emotions = df1['Emotion'].tolist()
```

```python
print(texts[0])
print(texts[1])
print(texts[2])
```

```python
tokenizer  = Tokenizer( oov_token="<OOV>")
tokenizer.fit_on_texts(texts)
```

```python
sequences = tokenizer.texts_to_sequences(texts)
print("Sequences:", sequences)
```

### Padding sequences is a technique used to ensure that all input sequences have the same length by adding a specific value (usually zeros) to the shorter sequences. Here's why padding is important:

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
max_len = max([len(seq) for seq in sequences])
```

```python
print(max_len)
```

```python
padded_sequences = pad_sequences(sequences, maxlen = max_len)
```

```python
# from sklearn.preprocessing import LabelEncoder
# label_encoder = LabelEncoder()
# labels = label_encoder.fit_transform(emotions)
# print(labels)
```

```python
#one_hot_labels = keras.utils.to_categorical(emotions)
encoded_df = pd.get_dummies(emotions,dtype = int)
print(encoded_df)
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(padded_sequences,  encoded_df,test_size = 0.2, random_state = 42)
```

* Now lets define nueral network architecture for to create a model

```python
from keras.models import Sequential
from keras.layers import Embedding, Flatten, Dense
```

```python
nlp_model = Sequential()
```

```python
nlp_model.add(Embedding(input_dim = len(tokenizer.word_index)+1,
                       output_dim =100, input_length = max_len))

nlp_model.add(Flatten())
nlp_model.add(Dense(units = 100, activation = 'relu'))
nlp_model.add(Dense(units = encoded_df.shape[1], activation = 'softmax'))
#nlp_model.add(Dense(units=encoded_df.shape[1], activation='softmax'))


nlp_model.compile(optimizer='adam', 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

nlp_model.fit(xtrain, ytrain, epochs=10, batch_size=32, validation_data=(xtest, ytest))
```

# Now we will test our model with some daily life examples

```python
print(df1['Text'].iloc[2000])
print(df1['Emotion'].iloc[2000])
```

```python
input_text = "i absolutely love this skinny fiber it is doing wonders for me and i feel fabulous"

# Preprocess the input text
input_sequence = tokenizer.texts_to_sequences([input_text])

padded_input_sequence = pad_sequences(input_sequence, maxlen=max_len)

prediction = nlp_model.predict(padded_input_sequence)

predicted_class_index = np.argmax(prediction, axis=-1)
emotion_label = encoded_df.columns[predicted_class_index]
print(emotion_label)

```
